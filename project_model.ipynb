```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split

# Load dataset
data_path = r"E:\Math 7243\lab1\project\Hemorrhage Segmentation Project\combined_matched_origins.csv"
data_df = pd.read_csv(data_path)

# Define the base directories for each Type
type_to_directory = {
    "epidural": r"E:\Math 7243\lab1\project\data\epidural",
    "intraparenchymal": r"E:\Math 7243\lab1\project\data\intraparenchymal",
    "intraventricular": r"E:\Math 7243\lab1\project\data\intraventricular",
    "multi": r"E:\Math 7243\lab1\project\data\multi",
    "normal": r"E:\Math 7243\lab1\project\data\normal",
    "subarachnoid": r"E:\Math 7243\lab1\project\data\subarachnoid",
    "subdural": r"E:\Math 7243\lab1\project\data\subdural"
}

# Define the 4 window subdirectories
windows = ["brain_bone_window", "brain_window", "max_contrast_window", "subdural_window"]

# Validate all windows
def validate_all_windows(row):
    base_dir = type_to_directory.get(row['Type'])
    if not base_dir:
        return False
    for window in windows:
        window_path = os.path.join(base_dir, window, row['Origin'])
        if not os.path.exists(window_path):
            return False
    return True

# Add 'Valid' column
data_df['Valid'] = data_df.apply(validate_all_windows, axis=1)
valid_data_df = data_df[data_df['Valid']].reset_index(drop=True)

# Map classes to integers
class_map = {cls: idx for idx, cls in enumerate(valid_data_df['Type'].unique())}
valid_data_df['Label'] = valid_data_df['Type'].map(class_map)

# Split dataset
train_df, temp_df = train_test_split(valid_data_df, test_size=0.3, stratify=valid_data_df['Label'], random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['Label'], random_state=42)

# Load image paths and labels
def load_image_paths_and_labels(df):
    image_paths = []
    labels = []
    for _, row in df.iterrows():
        for window in windows:
            path = os.path.join(type_to_directory[row['Type']], window, row['Origin'])
            image_paths.append(path)
            labels.append(row['Label'])
    return np.array(image_paths), np.array(labels)

train_paths, train_labels = load_image_paths_and_labels(train_df)
val_paths, val_labels = load_image_paths_and_labels(val_df)
test_paths, test_labels = load_image_paths_and_labels(test_df)

# Prepare DataFrame for generators
def prepare_dataframe(paths, labels, class_map):
    class_names = {v: k for k, v in class_map.items()}  # Reverse mapping: int -> class name
    df = pd.DataFrame({
        "filename": paths,
        "class": [class_names[label] for label in labels]
    })
    return df

train_df_gen = prepare_dataframe(train_paths, train_labels, class_map)
val_df_gen = prepare_dataframe(val_paths, val_labels, class_map)
test_df_gen = prepare_dataframe(test_paths, test_labels, class_map)

# Data augmentation
image_size = 224  # Image size for MobileNetV2
datagen = ImageDataGenerator(
    rescale=1.0/255.0,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest"
)

train_gen = datagen.flow_from_dataframe(
    dataframe=train_df_gen,
    x_col="filename",
    y_col="class",
    target_size=(image_size, image_size),
    batch_size=32,
    class_mode="categorical"
)

val_gen = datagen.flow_from_dataframe(
    dataframe=val_df_gen,
    x_col="filename",
    y_col="class",
    target_size=(image_size, image_size),
    batch_size=32,
    class_mode="categorical"
)

# Load MobileNetV2
base_model = MobileNetV2(weights="imagenet", include_top=False, input_shape=(image_size, image_size, 3))
base_model.trainable = False  # Freeze the base model layers

# Add custom layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(len(class_map), activation='softmax')(x)

# Build and compile model
model = Model(inputs=base_model.input, outputs=output)
model.compile(optimizer=Adam(learning_rate=1e-4), loss="categorical_crossentropy", metrics=["accuracy"])

# Train model
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10,  # Reduced epochs for faster training
    steps_per_epoch=len(train_gen),
    validation_steps=len(val_gen)
)

# Save model
model.save("mobilenetv2_hemorrhage_classifier.h5")

# Test generator
test_datagen = ImageDataGenerator(rescale=1.0/255.0)
test_gen = test_datagen.flow_from_dataframe(
    dataframe=test_df_gen,
    x_col="filename",
    y_col="class",
    target_size=(image_size, image_size),
    batch_size=32,
    class_mode="categorical",
    shuffle=False
)

# Evaluate model
test_loss, test_acc = model.evaluate(test_gen)
print(f"Test Accuracy: {test_acc * 100:.2f}%")
```

```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB4
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split

# Load dataset
data_path = r"E:\Math 7243\lab1\project\Hemorrhage Segmentation Project\combined_matched_origins.csv"
data_df = pd.read_csv(data_path)

# Define the base directories for each Type
type_to_directory = {
    "epidural": r"E:\Math 7243\lab1\project\data\epidural\epidural",
    "intraparenchymal": r"E:\Math 7243\lab1\project\data\intraparenchymal\intraparenchymal",
    "intraventricular": r"E:\Math 7243\lab1\project\data\intraventricular\intraventricular",
    "multi": r"E:\Math 7243\lab1\project\data\multi\multi",
    "normal": r"E:\Math 7243\lab1\project\data\normal\normal",
    "subarachnoid": r"E:\Math 7243\lab1\project\data\subarachnoid\subarachnoid",
    "subdural": r"E:\Math 7243\lab1\project\data\subdural\subdural"
}

# Define the 4 window subdirectories
windows = ["brain_bone_window", "brain_window", "max_contrast_window", "subdural_window"]

# Validate all windows
def validate_all_windows(row):
    base_dir = type_to_directory.get(row['Type'])
    if not base_dir:
        return False
    for window in windows:
        window_path = os.path.join(base_dir, window, row['Origin'])
        if not os.path.exists(window_path):
            return False
    return True

# Add 'Valid' column
data_df['Valid'] = data_df.apply(validate_all_windows, axis=1)
valid_data_df = data_df[data_df['Valid']].reset_index(drop=True)

# Map classes to integers
class_map = {cls: idx for idx, cls in enumerate(valid_data_df['Type'].unique())}
valid_data_df['Label'] = valid_data_df['Type'].map(class_map)

# Split dataset
train_df, temp_df = train_test_split(valid_data_df, test_size=0.3, stratify=valid_data_df['Label'], random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['Label'], random_state=42)

# Prepare DataFrame for generators
def prepare_dataframe(df):
    df['filename'] = df.apply(
        lambda row: os.path.join(type_to_directory[row['Type']], windows[0], row['Origin']), axis=1
    )
    df['class'] = df['Type']
    return df[['filename', 'class']]

train_df_gen = prepare_dataframe(train_df)
val_df_gen = prepare_dataframe(val_df)
test_df_gen = prepare_dataframe(test_df)

# Data augmentation
image_size = 380  # Input size for EfficientNetB4
datagen = ImageDataGenerator(
    rescale=1.0/255.0,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest"
)

train_gen = datagen.flow_from_dataframe(
    dataframe=train_df_gen,
    x_col="filename",
    y_col="class",
    target_size=(image_size, image_size),
    batch_size=32,
    class_mode="categorical"
)

val_gen = datagen.flow_from_dataframe(
    dataframe=val_df_gen,
    x_col="filename",
    y_col="class",
    target_size=(image_size, image_size),
    batch_size=32,
    class_mode="categorical"
)

# Load EfficientNetB4
base_model = EfficientNetB4(weights="imagenet", include_top=False, input_shape=(image_size, image_size, 3))
base_model.trainable = False  # Freeze the base model layers

# Add custom layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(len(class_map), activation='softmax')(x)

# Build and compile model
model = Model(inputs=base_model.input, outputs=output)
model.compile(optimizer=Adam(learning_rate=0.001), loss="categorical_crossentropy", metrics=["accuracy"])

# Train model
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=5,
    steps_per_epoch=train_gen.n // train_gen.batch_size,
    validation_steps=val_gen.n // val_gen.batch_size
)

# Save model
model.save("efficientnetb4_hemorrhage_classifier.h5")

# Test generator
test_datagen = ImageDataGenerator(rescale=1.0/255.0)
test_gen = test_datagen.flow_from_dataframe(
    dataframe=test_df_gen,
    x_col="filename",
    y_col="class",
    target_size=(image_size, image_size),
    batch_size=32,
    class_mode="categorical",
    shuffle=False
)

# Evaluate model
test_loss, test_acc = model.evaluate(test_gen)
print(f"Test Accuracy: {test_acc * 100:.2f}%")
```

```python

import cv2
import numpy as np

# Function to load and combine 4 windows into a single 4-channel image
def load_4_channel_image(base_dir, origin):
    windows = ["brain_bone_window", "brain_window", "max_contrast_window", "subdural_window"]
    channels = []
    for window in windows:
        path = os.path.join(base_dir, window, origin)
        image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)  # Load as grayscale
        if image is None:
            raise FileNotFoundError(f"File not found: {path}")
        channels.append(image)
    return np.stack(channels, axis=-1)  # Stack along the channel dimension

# Test on one sample
base_dir = r"E:\Math 7243\lab1\project\data\epidural\epidural"
origin = "example_image.png"  # Replace with a valid file
four_channel_image = load_4_channel_image(base_dir, origin)
print("4-Channel Image Shape:", four_channel_image.shape)
```

```python
class FourChannelDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, dataframe, batch_size, image_size, class_map):
        self.dataframe = dataframe
        self.batch_size = batch_size
        self.image_size = image_size
        self.class_map = class_map

    def __len__(self):
        return len(self.dataframe) // self.batch_size

    def __getitem__(self, index):
        batch = self.dataframe.iloc[index * self.batch_size:(index + 1) * self.batch_size]
        images = []
        labels = []
        for _, row in batch.iterrows():
            base_dir = type_to_directory[row['Type']]
            image = load_4_channel_image(base_dir, row['Origin'])
            image = cv2.resize(image, (self.image_size, self.image_size))  # Resize to desired size
            images.append(image)
            labels.append(self.class_map[row['Type']])
        return np.array(images), tf.keras.utils.to_categorical(labels, num_classes=len(self.class_map))
```

```python
from tensorflow.keras.applications import EfficientNetB4
from tensorflow.keras.layers import Input

# Input layer with 4 channels
input_shape = (380, 380, 4)
inputs = Input(shape=input_shape)

# Use EfficientNetB4
base_model = EfficientNetB4(weights=None, include_top=False, input_tensor=inputs)  # No pretrained weights

# Add custom layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(len(class_map), activation='softmax')(x)

# Build model
model = Model(inputs=inputs, outputs=output)

# Compile the model
model.compile(optimizer=Adam(learning_rate=1e-4), loss="categorical_crossentropy", metrics=["accuracy"])
```
